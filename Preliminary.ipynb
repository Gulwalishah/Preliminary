```python
import matplotlib.pyplot as plt
import numpy as np

# Histogram params
a = 0
b = 10
n = 5
x = np.linspace(a, b, n+1)
z = np.linspace(1, 10, n)

# Convert x to indices
x_indices = np.digitize(x, x)-1

# Linear spline
def linear_spline(x, z, x_indices):
  S = np.zeros_like(x)
  for i in range(n):
      idx1 = x_indices[i]
      idx2 = x_indices[i+1]
      S[idx1:idx2] = (z[i] + 0*x[idx1:idx2])/(1+0*x[idx1:idx2])
  return S

S = linear_spline(x, z, x_indices)
plt.plot(x, S)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Define histogram parameters
a = 0
b = 10
n = 5
x = np.linspace(a, b, n+1)
z = np.linspace(1, 10, n)

# Plot histogram
plt.bar(x[:-1], z, width=np.diff(x), align='edge')
plt.xlabel('x')
plt.ylabel('z')
plt.title('Histogram')
plt.show()

# Define linear spline function
def linear_spline(x, z, x_values):
    S = np.zeros_like(x)
    for i in range(n):
        S[x_values[i]:x_values[i+1]] = (z[i] + 0*x[x_values[i]:x_values[i+1]])/(1+0*x[x_values[i]:x_values[i+1]])
    return S

# Plot linear spline
S = linear_spline(x, z, x)
plt.plot(x, S)
plt.show()
```
